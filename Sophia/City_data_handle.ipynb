```python
import os 
import csv
import pandas as pd
import numpy as np
```

```python
file_name_sale = 'Sale_Prices_City.csv'
file_name_region = 'state_region.csv'
file_name_gymratio = 'city_gym_ratio.csv'


gym=pd.read_csv(file_name_gymratio)

region = pd.read_csv(file_name_region)

```

```python
# to remove space from raw data
file_name_loc = 'city_loc.csv'
city_loc = pd.read_csv(file_name_loc)
city_loc['area'] = city_loc['area'].str.strip()
city_loc.head()
city_loc.to_csv(file_name_loc)

city_loc.head()
```

```python
# to add region to cities
region_map = region[['State','State Code','Region']]
region_map.head()
```

```python
ratio_region = pd.DataFrame.merge(gym, region_map, left_on='state', how='inner', right_on='State Code', right_index=False)
ratio_region['State'] = ratio_region['State'].str.strip()
ratio_region['State Code'] = ratio_region['State Code'].str.strip()
ratio_region['Region'] = ratio_region['Region'].str.strip()
ratio_region.head()
```

```python
#ratio_region[['city','State','ratio%','Region']].to_csv('gym_ratio_region.csv', index=False)
```

```python
combined_df = pd.merge(ratio_region, city_loc, how='inner', left_on='city', right_on='area')
combined_df.head(2)
```

```python
columns = ['city','State','ratio%','Region','latitude','longitude','spending','weekly wage','State Code']
city_list = combined_df[columns]

city_list['city'] = city_list['city'].str.strip()
city_list['State'] = city_list['State'].str.strip()

city_list.to_csv('city_list.csv',index=False)
```

```python
city_loc = pd.read_csv(file_name_loc)
city_loc['area'] = city_loc['area'].str.strip()
city_loc.head()
city_loc.to_csv(file_name_loc)
```
